```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 7, 5]
y = [2, 4, 9, 19, 10]

plt.plot(x, y, marker='o', linestyle='-', color='m', label='Data Points')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.show()
plt.scatter(x, y, color='r', label='Scatter Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [4, 7, 1, 8]
})

sns.barplot(x='Category', y='Values', data=data)
plt.title('Seaborn Barplot')
plt.show()
```

```python
df = pd.read_csv("/home/saida/IdeaProjects/data-science-project/data/titanic.csv")

# Histogram of ages
plt.hist(df['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()

# Survival by gender
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival Count by Gender')
plt.show()
```
