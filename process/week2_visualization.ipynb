```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 7, 5]
y = [2, 4, 9, 19, 10]

plt.plot(x, y, marker='o', linestyle='-', color='m', label='Data Points')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.show()
plt.scatter(x, y, color='r', label='Scatter Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
